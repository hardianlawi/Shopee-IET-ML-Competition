```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix

%matplotlib inline
```

```python
labels = [
    "BabyBibs", 
    "BabyHat", 
    "BabyPants", 
    "BabyShirt", 
    "PackageFart", 
    "womanshirtsleeve", 
    "womencasualshoes", 
    "womenchiffontop", 
    "womendollshoes", 
    "womenknittedtop", 
    "womenlazyshoes", 
    "womenlongsleevetop", 
    "womenpeashoes", 
    "womenplussizedtop", 
    "womenpointedflatshoes", 
    "womensleevelesstop", 
    "womenstripedtop", 
    "wrapsnslings",
]
```

```python
InceptionV3 = pd.read_csv("./outputs/val/InceptionV3_val.csv")
InceptionResNetV2 = pd.read_csv("./outputs/val/InceptionResNetV2_val.csv")
Xception = pd.read_csv("./outputs/val/Xception_val.csv")
ResNet50 = pd.read_csv("./outputs/val/ResNet50_val.csv")
```

```python
# InceptionV3

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(16, 12))

y_true = InceptionV3.category_id
y_pred = np.argmax(InceptionV3.drop("category_id", axis=1).as_matrix(), axis=-1)

sns.heatmap(confusion_matrix(y_true, y_pred), cmap="YlGnBu", ax=ax[0][0], annot=True, fmt="d")
ax[0][0].set_title("InceptionV3")

# InceptionResNetV2

y_true = InceptionResNetV2.category_id
y_pred = np.argmax(InceptionResNetV2.drop("category_id", axis=1).as_matrix(), axis=-1)

sns.heatmap(confusion_matrix(y_true, y_pred), cmap="YlGnBu", ax=ax[0][1], annot=True, fmt="d")
ax[0][1].set_title("InceptionResNetV2")

# Xception

y_true = Xception.category_id
y_pred = np.argmax(Xception.drop("category_id", axis=1).as_matrix(), axis=-1)

sns.heatmap(confusion_matrix(y_true, y_pred), cmap="YlGnBu", ax=ax[1][0], annot=True, fmt="d")
ax[1][0].set_title("Xception")

# ResNet50

y_true = ResNet50.category_id
y_pred = np.argmax(ResNet50.drop("category_id", axis=1).as_matrix(), axis=-1)

sns.heatmap(confusion_matrix(y_true, y_pred), cmap="YlGnBu", ax=ax[1][1], annot=True, fmt="d")
ax[1][1].set_title("ResNet50")

plt.tight_layout()
```

```python
# InceptionV3

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(16, 12))

y_true = InceptionV3.category_id
y_pred = np.argmax(InceptionV3.drop("category_id", axis=1).as_matrix(), axis=-1)

preds = []
for idx, ser in InceptionV3.loc[y_true != y_pred].iterrows():
    preds.append(ser[y_pred[idx]])

sns.distplot(preds, ax=ax[0][0])
ax[0][0].set_title("InceptionV3")
ax[0][0].set_ylabel("Frequency")
ax[0][0].set_xlabel("Confidence")

# InceptionResNetV2

y_true = InceptionResNetV2.category_id
y_pred = np.argmax(InceptionResNetV2.drop("category_id", axis=1).as_matrix(), axis=-1)

preds = []
for idx, ser in InceptionResNetV2.loc[y_true != y_pred].iterrows():
    preds.append(ser[y_pred[idx]])

sns.distplot(preds, ax=ax[0][1])
ax[0][1].set_title("InceptionResNetV2")
ax[0][1].set_ylabel("Frequency")
ax[0][1].set_xlabel("Confidence")

# Xception

y_true = Xception.category_id
y_pred = np.argmax(Xception.drop("category_id", axis=1).as_matrix(), axis=-1)

preds = []
for idx, ser in Xception.loc[y_true != y_pred].iterrows():
    preds.append(ser[y_pred[idx]])

sns.distplot(preds, ax=ax[1][0])
ax[1][0].set_title("Xception")
ax[1][0].set_ylabel("Frequency")
ax[1][0].set_xlabel("Confidence")

# ResNet50

y_true = ResNet50.category_id
y_pred = np.argmax(ResNet50.drop("category_id", axis=1).as_matrix(), axis=-1)

preds = []
for idx, ser in ResNet50.loc[y_true != y_pred].iterrows():
    preds.append(ser[y_pred[idx]])

sns.distplot(preds, ax=ax[1][1])
ax[1][1].set_title("ResNet50")
ax[1][1].set_ylabel("Frequency")
ax[1][1].set_xlabel("Confidence")

plt.tight_layout()
```
